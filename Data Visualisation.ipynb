```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('instahyre_data.csv')
df
```

Histogram of Founded Year


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Founded'].dropna(), bins=20, kde=True)
plt.title('Distribution of Founded Year')
plt.xlabel('Founded Year')
plt.ylabel('Frequency')
plt.show()
```

Box Plot on Employee count.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Employees', y='Location', data=df)
plt.title('Box Plot of Employee Count by Location')
plt.xlabel('Number of Employees')
plt.ylabel('Location')
plt.show()
```

Pair Plot of Founded Year and Employee Count

```python
sns.pairplot(df[['Founded', 'Employees']])
plt.title('Pair Plot of Founded Year and Employee Count')
plt.show()
```

Heatmap of correlation matrix

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

Word Cloud of Skills

```python
from wordcloud import WordCloud

skills_text = ' '.join(df['Skills'].explode().dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(skills_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Bar Chart of Job Distribution by Location

```python
plt.figure(figsize=(12, 6))
location_counts = df['Location'].value_counts().head(10)
location_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Job Locations')
plt.xlabel('Location')
plt.ylabel('Number of Jobs')
plt.show()
```

Bar Chart of top 10 Mentioned Skills

```python
plt.figure(figsize=(12, 6))
top_skills = df['Skills'].explode().value_counts().head(10)
top_skills.plot(kind='bar', color='green')
plt.title('Top 10 Most Mentioned Skills')
plt.xlabel('Skill')
plt.ylabel('Number of Mentions')
plt.show()
```

Scatter Plot : Founded Vs Employees

```python
plt.scatter(df['Founded'], df['Employees'], color='skyblue', alpha=0.5)
plt.title('Scatter Plot: Founded vs Employees')
plt.xlabel('Founded Year')
plt.ylabel('Number of Employees')
plt.show()
```
